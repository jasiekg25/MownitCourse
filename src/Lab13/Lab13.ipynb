# Zadanie 1
Zaimplementuj generator liniowy kongruentny wg wzoru:
$$X_{i+1}=(aX_i+c)\ mod\ m$$
Generator ma zwracać wartości w zakresie [0...1], wartości startowe wg własnego wyboru.
 - Przetestuj jego działanie dla różnych zestawów parametrów:

        |     a         |   c           | m          |
        | ------------- |:-------------:| ---------: |
        |  1103515245   | 12345         | $$2^{32}$$ |
        | 1229          | 1             | $$2048$$   |
        | 1597          | 51749         | $$24494$$  |
        | $2^{16}+3$    | 0             | $ 2^{31}$  |
 - Zwizualizuj rozkład generowanych liczby (histogram lub lepiej chmura punktów 2-D lub 3D ($X_i, X_{i+1}$), ($X_i, X_{i+1}, X_{i+2}$)).
 Oceń jakość generowanych rozkładów.

```python
import matplotlib.pyplot as plt
import numpy as np
array = []


def random_number(a, c, m, x0):
    return (a*x0 + c) % m


def n_random_numbers(a, c, m, x0, n):
    array = []
    for _ in range(n):
        x0 = random_number(a, c, m, x0)
        array.append(x0/m)
    return array
```

```python
tmp_array = n_random_numbers(1103515245, 12345, 2**32, 1000, 30)
array += tmp_array
print(tmp_array)
```

```python
tmp_array = n_random_numbers(1229, 1, 2048, 1000, 30)
array += tmp_array
print(tmp_array)
```

```python
tmp_array = n_random_numbers(1597, 51749, 24494, 1000, 30)
array += tmp_array
print(tmp_array)
```

```python
tmp_array = n_random_numbers(2**16 + 3, 0, 2**31, 1000, 30)
array += tmp_array
print(tmp_array)
```

```python
x = [array[i] for i in range(len(array)) if i % 2 == 0]
y = [array[i] for i in range(len(array)) if i % 2 == 1]

plt.figure(figsize=(16, 12), dpi=100)
plt.scatter(x, y, alpha=0.7,cmap='inferno')
plt.show()
```

# Zadanie 3
Zaimplementuj metodą całkowania Monte Carlo i zastosuj ją wobec następujących całek:
$$ \int_1^2 \frac{1}{x^2}\ dx $$ ,
$$ \int_1^6 \frac{1}{\sqrt{x^5+8}}\ dx $$ ,
$$ \int_1^{22} x^2+2x\ dx $$ ,


```python
def monte_carlo(fun, a, b, N):
    xrand = [np.random.uniform(a, b) for _ in range(N)]
    result = 0.0
    for i in range(N):
        result += fun(xrand[i])
    return (b-a)/float(N)*result
```

```python
# rozwiązuję dla 4 różnych liczb próbek (dokładności)
f1 = lambda x: 1 / (x**2)  # [1,2]
print(monte_carlo(f1, 1, 2, 100))
print(monte_carlo(f1, 1, 2, 1000))
print(monte_carlo(f1, 1, 2, 10000))
print(monte_carlo(f1, 1, 2, 100000))
```

```python
f2 = lambda x: 1 / np.sqrt(x**5+8)  # [1,6]
print(monte_carlo(f2, 1, 6, 100))
print(monte_carlo(f2, 1, 6, 1000))
print(monte_carlo(f2, 1, 6, 10000))
print(monte_carlo(f2, 1, 6, 100000))
```

```python
f3 = lambda x: x**2 + 2*x  # [1,22]
print(monte_carlo(f3, 1, 22, 100))
print(monte_carlo(f3, 1, 22, 1000))
print(monte_carlo(f3, 1, 22, 10000))
print(monte_carlo(f3, 1, 22, 100000))
```

 ## Wyjaśnienie podejścia do zadania

 Znamy zależność:

  $\frac{n}{N} = \frac{V1}{V2}$

    n - ilość punktów wylosowanych, zawierających się w figurze,
    N - ilość punktów wylosowanych, zawierających się w sześcianie który ogranicza figurę,
    V1 - objętości figury,
    V2 - objętości sześcianu, który ogranicza figure

### Kula o promieniu 5

```python
def condition1(x, y, z, R):
    return x ** 2 + y ** 2 + z ** 2 <= R ** 2

def sphere(R, N):
    counter = 0
    for _ in range(N):
        point = np.random.uniform(-R, R), np.random.uniform(-R, R), np.random.uniform(-R, R)
        if condition1(*point, R):
            counter += 1
    return (counter / N) * ((2 * R) ** 3)

print("V rzeczywiste wynosi: ", 4*3.14*125/3 )
print("R = 5 \tN=1000 \t\tV=",sphere(5, 1000))
print("R = 5 \tN=10000 \tV=",sphere(5, 10000))
print("R = 5 \tN=100000 \tV=",sphere(5, 100000))
print("R = 5 \tN=1000000 \tV=",sphere(5, 1000000))
```

### Stożek o wysokości 10 i promieniu podstawy 10

```python
def condition2(x, y, z, R, h):
    return x ** 2 + y ** 2 <= ((z*R)/h) ** 2

def cone(R, h, N):
    counter = 0
    for _ in range(N):
        point = np.random.uniform(-R, R), np.random.uniform(-R, R), np.random.uniform(0, h)
        if condition2(*point, R, h):
            counter += 1
    return (counter / N) * (h * (2 * R) ** 2)
print("V rzeczywiste wynosi: \t\t", 3.14*100*10/3 )
print("H = 10 \tR = 10 \tN=1000 \t\tV=",cone(10, 10, 1000))
print("H = 10 \tR = 10 \tN=10000 \tV=",cone(10, 10, 10000))
print("H = 10 \tR = 10 \tN=100000 \tV=",cone(10, 10, 100000))
print("H = 10 \tR = 10 \tN=1000000 \tV=",cone(10, 10, 1000000))
```
